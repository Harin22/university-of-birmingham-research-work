```python
import torch, cv2, numpy as np, pandas as pd
from detectron2 import model_zoo
from detectron2.engine import DefaultPredictor
from detectron2.config import get_cfg
from detectron2.data import MetadataCatalog

def calculate_bubble_measurements(mask, bbox):
    measurements = {}
    x1, y1, x2, y2 = bbox
    measurements['bbox_width'] = float(x2 - x1)
    measurements['bbox_height'] = float(y2 - y1)

    mask_area = np.sum(mask)
    measurements['mask_area_pixels'] = float(mask_area)
    measurements['equivalent_radius'] = float(np.sqrt(mask_area / np.pi))  #measure radius

    mask_uint8 = mask.astype(np.uint8) * 255
    contours, _ = cv2.findContours(mask_uint8, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    if contours:
        largest_contour = max(contours, key=cv2.contourArea)
        if len(largest_contour) >= 5:
            ellipse = cv2.fitEllipse(largest_contour)
            center, axes, angle = ellipse
            measurements['major_axis'] = float(max(axes))
            measurements['minor_axis'] = float(min(axes))
            measurements['aspect_ratio'] = float(max(axes) / min(axes))
        measurements['perimeter'] = float(cv2.arcLength(largest_contour, True))

    return measurements

cfg = get_cfg()
cfg.merge_from_file(model_zoo.get_config_file("COCO-InstanceSegmentation/mask_rcnn_R_50_FPN_3x.yaml"))
cfg.MODEL.ROI_HEADS.NUM_CLASSES = 2
cfg.MODEL.WEIGHTS = "./bubble_model_output/model_final.pth"
cfg.MODEL.ROI_HEADS.SCORE_THRESH_TEST = 0.5

bubble_metadata = MetadataCatalog.get("bubble_measurement")
bubble_metadata.thing_classes = ["Bubbles-IFQb", "taylor bubble"]

predictor = DefaultPredictor(cfg)

image = cv2.imread("/content/drive/MyDrive/Uni Research/sample_image.png")
outputs = predictor(image)
instances = outputs["instances"].to("cpu")

bubble_data = []
boxes = instances.pred_boxes.tensor.numpy()
masks = instances.pred_masks.numpy()
scores = instances.scores.numpy()
classes = instances.pred_classes.numpy()

print(" TAYLOR BUBBLE MEASUREMENTS:")
print("="*50)

for i in range(len(instances)):
    measurements = calculate_bubble_measurements(masks[i], boxes[i])
    bubble_type = bubble_metadata.thing_classes[classes[i]]

    bubble_info = {
        'bubble_id': i + 1,
        'type': bubble_type,
        'confidence': scores[i] * 100,
        **measurements
    }
    bubble_data.append(bubble_info)

    print(f"\n BUBBLE #{i+1} - {bubble_type}")
    print(f"    Length: {measurements['bbox_width']:.1f} pixels")
    print(f"    Height: {measurements['bbox_height']:.1f} pixels")
    print(f"    Radius: {measurements['equivalent_radius']:.1f} pixels")
    print(f"    Area: {measurements['mask_area_pixels']:.0f} pixels²")
    if 'major_axis' in measurements:
        print(f"    Major Axis: {measurements['major_axis']:.1f} pixels")
        print(f"    Aspect Ratio: {measurements['aspect_ratio']:.2f}")

df = pd.DataFrame(bubble_data)
df.to_csv("taylor_bubble_measurements.csv", index=False)
print(f"\n Measurements saved to CSV file!")
print(f" Summary: {len(instances)} bubbles analyzed")
```

```python

import torch, cv2, numpy as np, pandas as pd
from detectron2 import model_zoo
from detectron2.engine import DefaultPredictor
from detectron2.config import get_cfg
from detectron2.data import MetadataCatalog
from google.colab.patches import cv2_imshow

def calculate_measurements(mask, bbox):
    x1, y1, x2, y2 = bbox
    length = float(x2 - x1)
    height = float(y2 - y1)
    area = float(np.sum(mask))
    radius = float(np.sqrt(area / np.pi))

    return {'length': length, 'height': height, 'area': area, 'radius': radius}

def create_super_clean_image(image, instances, metadata):
    result = image.copy()

    if len(instances) == 0:
        return result

    boxes = instances.pred_boxes.tensor.numpy()
    masks = instances.pred_masks.numpy()
    scores = instances.scores.numpy()
    classes = instances.pred_classes.numpy()

    colors = [(0, 255, 0), (255, 100, 0)]  #Bright green, Bright orange

    for i in range(len(instances)):
        measurements = calculate_measurements(masks[i], boxes[i])
        color = colors[classes[i]]

        x1, y1, x2, y2 = boxes[i].astype(int)
        cv2.rectangle(result, (x1, y1), (x2, y2), color, 3)

        #bubble ID number
        cv2.putText(result, f"{i+1}", (x1-25, y1-10),
                   cv2.FONT_HERSHEY_SIMPLEX, 1.2, color, 3)

        mask_overlay = np.zeros_like(result)
        mask_overlay[masks[i]] = color
        result = cv2.addWeighted(result, 0.85, mask_overlay, 0.3, 0)

        text_x = x1 + 10
        text_y = y1 + 30

        measurements_text = [
            f"L:{measurements['length']:.0f}px",
            f"H:{measurements['height']:.0f}px",
            f"R:{measurements['radius']:.0f}px"
        ]

        for j, text in enumerate(measurements_text):
            y_pos = text_y + (j * 25)

            (text_w, text_h), _ = cv2.getTextSize(text, cv2.FONT_HERSHEY_SIMPLEX, 0.7, 2)
            cv2.rectangle(result, (text_x-5, y_pos-text_h-5),
                         (text_x+text_w+5, y_pos+5), (0, 0, 0), -1)

            cv2.putText(result, text, (text_x, y_pos),
                       cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2)

    cv2.rectangle(result, (0, 0), (result.shape[1], 50), (0, 0, 0), -1)
    cv2.putText(result, f"TAYLOR BUBBLE DETECTION - {len(instances)} BUBBLES FOUND",
               (10, 35), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 2)

    return result

def save_measurements_csv(instances, metadata):
    data = []
    boxes = instances.pred_boxes.tensor.numpy()
    masks = instances.pred_masks.numpy()
    scores = instances.scores.numpy()
    classes = instances.pred_classes.numpy()

    for i in range(len(instances)):
        measurements = calculate_measurements(masks[i], boxes[i])
        data.append({
            'Bubble_ID': i + 1,
            'Type': metadata.thing_classes[classes[i]],
            'Confidence_%': round(scores[i] * 100, 1),
            'Length_px': round(measurements['length'], 1),
            'Height_px': round(measurements['height'], 1),
            'Radius_px': round(measurements['radius'], 1),
            'Area_px2': round(measurements['area'], 0)
        })

    df = pd.DataFrame(data)
    df.to_csv("bubble_measurements.csv", index=False)

    print(f"💾 CSV SAVED: bubble_measurements.csv")
    print(f"📊 Data for {len(data)} bubbles")
    print("\n📈 SUMMARY:")
    print(f"   Average Length: {df['Length_px'].mean():.1f} px")
    print(f"   Average Radius: {df['Radius_px'].mean():.1f} px")
    print("\n📋 FULL DATA:")
    print(df.to_string(index=False))

    return df

cfg = get_cfg()
cfg.merge_from_file(model_zoo.get_config_file("COCO-InstanceSegmentation/mask_rcnn_R_50_FPN_3x.yaml"))
cfg.MODEL.ROI_HEADS.NUM_CLASSES = 2
cfg.MODEL.WEIGHTS = "./bubble_model_output/model_final.pth"
cfg.MODEL.ROI_HEADS.SCORE_THRESH_TEST = 0.5

metadata = MetadataCatalog.get("bubble_simple")
metadata.thing_classes = ["Bubbles-IFQb", "taylor bubble"]
predictor = DefaultPredictor(cfg)

image = cv2.imread("/content/drive/MyDrive/Uni Research/sample_image.png")
outputs = predictor(image)
instances = outputs["instances"].to("cpu")

print(f"🎯 FOUND {len(instances)} TAYLOR BUBBLES!")

clean_result = create_super_clean_image(image, instances, metadata)
cv2_imshow(clean_result)
cv2.imwrite("super_clean_bubbles.jpg", clean_result)

#Save CSV
df = save_measurements_csv(instances, metadata)

print("\n✅ MUCH CLEANER! Perfect for presentations!")
```

```python
import pandas as pd

def save_bubble_measurements_to_csv(instances, metadata, image_name="sample_image"):
    """Save all bubble measurements to CSV file"""
    print("📊 Extracting measurements for CSV export...")

    if len(instances) == 0:
        print("❌ No bubbles detected to save")
        return None

    boxes = instances.pred_boxes.tensor.numpy()
    masks = instances.pred_masks.numpy()
    scores = instances.scores.numpy()
    classes = instances.pred_classes.numpy()

    bubble_data = []

    for i in range(len(instances)):
        # Get measurements
        x1, y1, x2, y2 = boxes[i]
        length_px = float(x2 - x1)
        height_px = float(y2 - y1)
        mask_area = float(np.sum(masks[i]))
        equivalent_radius = float(np.sqrt(mask_area / np.pi))
        aspect_ratio = length_px / height_px if height_px > 0 else 0

        # Get bubble info
        bubble_type = metadata.thing_classes[classes[i]]
        confidence = float(scores[i] * 100)

        # Create record
        bubble_record = {
            'Bubble_ID': i + 1,
            'Bubble_Type': bubble_type,
            'Confidence_Percent': round(confidence, 1),
            'Length_Pixels': round(length_px, 2),
            'Height_Pixels': round(height_px, 2),
            'Equivalent_Radius_Pixels': round(equivalent_radius, 2),
            'Area_Pixels_Squared': round(mask_area, 0),
            'Aspect_Ratio': round(aspect_ratio, 3)
        }

        bubble_data.append(bubble_record)
        print(f"✅ Bubble {i+1}: {bubble_type} - L:{length_px:.1f} H:{height_px:.1f} R:{equivalent_radius:.1f}")

    #saving to CSV
    df = pd.DataFrame(bubble_data)
    filename = "taylor_bubble_measurements.csv"     #saving to CSV
    df.to_csv(filename, index=False)

    print(f"\n💾 SUCCESS! CSV saved as: {filename}")
    print(f"📋 Exported {len(bubble_data)} bubble measurements")

    print("\n📈 SUMMARY STATISTICS:")
    print(f"   Average Length: {df['Length_Pixels'].mean():.1f} pixels")
    print(f"   Average Height: {df['Height_Pixels'].mean():.1f} pixels")
    print(f"   Average Radius: {df['Equivalent_Radius_Pixels'].mean():.1f} pixels")

    print("\n📊 DATA TABLE:")
    print(df.to_string(index=False))

    return df

measurements_df = save_bubble_measurements_to_csv(instances, bubble_metadata, "sample_image.png")
```

```python

import torch, cv2, numpy as np, pandas as pd, time
from detectron2 import model_zoo
from detectron2.engine import DefaultPredictor
from detectron2.config import get_cfg
from detectron2.data import MetadataCatalog
from pathlib import Path

class BubbleVideoAnalyzer:
    def __init__(self, model_path):
        cfg = get_cfg()
        cfg.merge_from_file(model_zoo.get_config_file("COCO-InstanceSegmentation/mask_rcnn_R_50_FPN_3x.yaml"))
        cfg.MODEL.ROI_HEADS.NUM_CLASSES = 2
        cfg.MODEL.WEIGHTS = model_path
        cfg.MODEL.ROI_HEADS.SCORE_THRESH_TEST = 0.5

        self.metadata = MetadataCatalog.get("bubble_video")
        self.metadata.thing_classes = ["Bubbles-IFQb", "taylor bubble"]
        self.predictor = DefaultPredictor(cfg)
        print(f" Model loaded: {model_path}")

    def calculate_measurements(self, mask, bbox, frame_num, timestamp):
        x1, y1, x2, y2 = bbox
        length_px = float(x2 - x1)
        height_px = float(y2 - y1)
        mask_area = float(np.sum(mask))
        equivalent_radius = float(np.sqrt(mask_area / np.pi))
        aspect_ratio = length_px / height_px if height_px > 0 else 0

        return {
            'frame_number': frame_num,
            'timestamp_seconds': round(timestamp, 3),
            'length_pixels': round(length_px, 2),
            'height_pixels': round(height_px, 2),
            'equivalent_radius_pixels': round(equivalent_radius, 2),
            'area_pixels': round(mask_area, 0),
            'aspect_ratio': round(aspect_ratio, 3)
        }

    def process_video(self, video_path, sample_rate=5):
        cap = cv2.VideoCapture(str(video_path))
        fps = cap.get(cv2.CAP_PROP_FPS) or 30
        total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
        duration = total_frames / fps

        print(f"\n PROCESSING: {Path(video_path).name}")
        print(f" Duration: {duration:.1f} seconds |  Frames: {total_frames}")

        all_measurements = []
        frame_count = 0
        processed_count = 0
        start_time = time.time()

        while True:
            ret, frame = cap.read()
            if not ret:
                break

            if frame_count % sample_rate == 0:
                timestamp = frame_count / fps
                outputs = self.predictor(frame)
                instances = outputs["instances"].to("cpu")

                if len(instances) > 0:
                    boxes = instances.pred_boxes.tensor.numpy()
                    masks = instances.pred_masks.numpy()
                    scores = instances.scores.numpy()
                    classes = instances.pred_classes.numpy()

                    for i in range(len(instances)):
                        bubble_type = self.metadata.thing_classes[classes[i]]
                        confidence = scores[i] * 100

                        measurements = self.calculate_measurements(
                            masks[i], boxes[i], frame_count, timestamp)

                        measurements.update({
                            'bubble_id_in_frame': i + 1,
                            'bubble_type': bubble_type,
                            'confidence_percent': round(confidence, 1),
                            'video_file': Path(video_path).name
                        })

                        all_measurements.append(measurements)

                processed_count += 1

                if processed_count % 100 == 0:
                    elapsed = time.time() - start_time
                    bubbles_found = len(all_measurements)
                    print(f" Processed {processed_count} frames |  Found {bubbles_found} bubbles")

            frame_count += 1

        cap.release()
        elapsed = time.time() - start_time

        print(f"\n COMPLETE! Processed {processed_count} frames in {elapsed:.1f}s")
        print(f" Total bubbles detected: {len(all_measurements)}")

        return all_measurements

    def save_results(self, measurements, video_name):
        if not measurements:
            print(" No measurements to save")
            return

        df = pd.DataFrame(measurements)
        csv_path = f"{video_name}_bubble_measurements.csv"
        df.to_csv(csv_path, index=False)

        print(f"\n ANALYSIS SUMMARY")
        print(f" Total bubbles: {len(df)}")
        print(f" Average radius: {df['equivalent_radius_pixels'].mean():.1f} pixels")
        print(f" Average area: {df['area_pixels'].mean():.0f} pixels²")
        print(f" Average aspect ratio: {df['aspect_ratio'].mean():.2f}")
        print(f" Saved to: {csv_path}")

        return df

MODEL_PATH = "./bubble_model_output/model_final.pth"
VIDEO_PATH = "/content/drive/MyDrive/Uni Research/Air_to_water_sample_vid.MOV"

print(" TAYLOR BUBBLE VIDEO ANALYSIS STARTING")
analyzer = BubbleVideoAnalyzer(MODEL_PATH)

measurements = analyzer.process_video(VIDEO_PATH, sample_rate=5)

results_df = analyzer.save_results(measurements, "Air_to_water_sample_vid")

print("\n inference COMPLETE!")
```

```python

import torch, cv2, numpy as np, pandas as pd, time
from detectron2 import model_zoo
from detectron2.engine import DefaultPredictor
from detectron2.config import get_cfg
from detectron2.data import MetadataCatalog
from google.colab import drive

class CompleteBubbleAnalyzer:
    def __init__(self, model_path):
        cfg = get_cfg()
        cfg.merge_from_file(model_zoo.get_config_file("COCO-InstanceSegmentation/mask_rcnn_R_50_FPN_3x.yaml"))
        cfg.MODEL.ROI_HEADS.NUM_CLASSES = 2
        cfg.MODEL.WEIGHTS = model_path
        cfg.MODEL.ROI_HEADS.SCORE_THRESH_TEST = 0.5

        self.metadata = MetadataCatalog.get("bubble_test_analysis")
        self.metadata.thing_classes = ["Bubbles-IFQb", "taylor bubble"]
        self.predictor = DefaultPredictor(cfg)
        self.measurements_data = []
        print(f" Model loaded: {model_path}")

    def calculate_measurements(self, mask, bbox, frame_num, timestamp):
        """Calculate detailed bubble measurements"""
        x1, y1, x2, y2 = bbox

        # Core measurements
        length_px = float(x2 - x1)
        height_px = float(y2 - y1)
        mask_area = float(np.sum(mask))
        equivalent_radius = float(np.sqrt(mask_area / np.pi))
        aspect_ratio = length_px / height_px if height_px > 0 else 0

        # Perimeter
        perimeter = 0
        try:
            contours, _ = cv2.findContours(mask.astype(np.uint8), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
            if contours:
                perimeter = float(cv2.arcLength(max(contours, key=cv2.contourArea), True))
        except: pass

        return {
            'frame_number': frame_num,
            'timestamp_seconds': round(timestamp, 3),
            'length_pixels': round(length_px, 2),
            'height_pixels': round(height_px, 2),
            'equivalent_radius_pixels': round(equivalent_radius, 2),
            'actual_area_pixels': round(mask_area, 0),
            'perimeter_pixels': round(perimeter, 2),
            'aspect_ratio': round(aspect_ratio, 3)
        }

    def draw_annotations(self, frame, instances, frame_num, timestamp):
        """Draw bubble annotations with measurements"""
        if len(instances) == 0:
            return frame

        result = frame.copy()
        boxes = instances.pred_boxes.tensor.numpy()
        masks = instances.pred_masks.numpy()
        scores = instances.scores.numpy()
        classes = instances.pred_classes.numpy()

        for i in range(len(instances)):
            bubble_type = self.metadata.thing_classes[classes[i]]
            confidence = scores[i] * 100
            color = (0, 255, 0) if classes[i] == 0 else (255, 100, 0)

            measurements = self.calculate_measurements(masks[i], boxes[i], frame_num, timestamp)

            record = measurements.copy()
            record.update({
                'bubble_id_in_frame': i + 1,
                'bubble_type': bubble_type,
                'confidence_percent': round(confidence, 1)
            })
            self.measurements_data.append(record)

            x1, y1, x2, y2 = boxes[i].astype(int)
            cv2.rectangle(result, (x1, y1), (x2, y2), color, 2)

            mask_overlay = np.zeros_like(result)
            mask_overlay[masks[i]] = color
            result = cv2.addWeighted(result, 0.85, mask_overlay, 0.25, 0)

            #Bubble ID
            cv2.putText(result, f"#{i+1}", (x1-25, y1-5),
                       cv2.FONT_HERSHEY_SIMPLEX, 0.8, color, 2)

            length = measurements['length_pixels']
            height = measurements['height_pixels']
            radius = measurements['equivalent_radius_pixels']

            texts = [f"L:{length:.1f}px", f"H:{height:.1f}px", f"R:{radius:.1f}px", f"{confidence:.0f}%"]

            for j, text in enumerate(texts):
                y_pos = y1 + 25 + (j * 16)
                (tw, th), _ = cv2.getTextSize(text, cv2.FONT_HERSHEY_SIMPLEX, 0.45, 1)
                cv2.rectangle(result, (x1+3, y_pos-th-2), (x1+tw+5, y_pos+2), (0,0,0), -1)
                cv2.putText(result, text, (x1+5, y_pos), cv2.FONT_HERSHEY_SIMPLEX, 0.45, (255,255,255), 1)

        info = f"Frame {frame_num} | Time: {timestamp:.1f}s | Bubbles: {len(instances)} | Model: 88.6% mAP"
        cv2.rectangle(result, (0, 0), (result.shape[1], 35), (0, 0, 0), -1)
        cv2.putText(result, info, (10, 25), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 2)

        return result

    def process_video(self, input_path, output_path, max_frames=300):
        """Process video with complete analysis"""
        drive.mount('/content/drive')

        cap = cv2.VideoCapture(str(input_path))
        fps = int(cap.get(cv2.CAP_PROP_FPS)) or 30
        width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
        height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
        total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

        print(f"\n🎥 PROCESSING: bubble_test_original_vid.mp4")
        print(f" {width}x{height} | {fps} FPS | {total_frames} frames | {total_frames/fps:.1f}s")

        #Output video
        fourcc = cv2.VideoWriter_fourcc(*'mp4v')
        out = cv2.VideoWriter(str(output_path), fourcc, fps, (width, height))

        frame_count = 0
        start_time = time.time()

        while frame_count < max_frames:
            ret, frame = cap.read()
            if not ret: break

            timestamp = frame_count / fps
            outputs = self.predictor(frame)
            instances = outputs["instances"].to("cpu")

            annotated = self.draw_annotations(frame, instances, frame_count, timestamp)
            out.write(annotated)

            frame_count += 1
            if frame_count % 50 == 0:
                elapsed = time.time() - start_time
                print(f" Frame {frame_count} | 🫧 {len(self.measurements_data)} bubbles |  {frame_count/elapsed:.1f} FPS")

        cap.release()
        out.release()

        print(f"\ VIDEO COMPLETE! {len(self.measurements_data)} bubbles detected")

    def save_csv(self, filename):
        """Save measurements to CSV"""
        df = pd.DataFrame(self.measurements_data)
        df.to_csv(filename, index=False)

        print(f"\n CSV SAVED: {filename}")
        print(f"SUMMARY STATS:")
        print(f"   Average Length: {df['length_pixels'].mean():.1f}px")
        print(f"   Average Height: {df['height_pixels'].mean():.1f}px")
        print(f"   Average Radius: {df['equivalent_radius_pixels'].mean():.1f}px")
        print(f"   Average Area: {df['actual_area_pixels'].mean():.0f}px²")

        return df

MODEL_PATH = "./bubble_model_output/model_final.pth"
INPUT_VIDEO = "/content/drive/MyDrive/Uni Research/bubble_test_original_vid.mp4"
OUTPUT_VIDEO = "/content/drive/MyDrive/bubble_test_analysis_results.mp4"
CSV_FILE = "/content/drive/MyDrive/bubble_test_measurements.csv"

print(" BUBBLE_TEST_ORIGINAL_VID.MP4 - COMPLETE ANALYSIS")
print("=" * 60)

analyzer = CompleteBubbleAnalyzer(MODEL_PATH)
analyzer.process_video(INPUT_VIDEO, OUTPUT_VIDEO, max_frames=300)  # ~10 seconds
results_df = analyzer.save_csv(CSV_FILE)

print("\n ANALYSIS COMPLETE!")
print("Results Video: Google Drive > bubble_test_analysis_results.mp4")
print(" Measurements CSV: Google Drive > bubble_test_measurements.csv")

```
